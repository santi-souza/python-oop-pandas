```python
class Producto:

    def __init__(self, codigo, nombre, precio, tipo): # SIEMPRE TIENE UN PARAMETRO INICIAL self.
        self.__codigo = codigo # Variable privada
        self.__nombre = nombre # Variable privada
        self.__precio = precio # Variable privada
        self.__tipo = tipo # Variable privada
        print("Persona creada con exito")

    # Los getters y setters son cuando tenemos variables privadas.
    # Getter 
    @property # Se denomina decorador, y le permite acceder a la variable codigo en este caso.
    def codigo(self):
        print("Getter del codigo")
        return self.__codigo

    # Setter. No se puede crear si no tengo el getter previamente definido. Sirve para modificar los datos de esa variable.
    @codigo.setter
    def codigo(self, nuevoCodigo):
        print("Setter del codigo")
        self.__codigo = nuevoCodigo

    @property
    def nombre(self):
        print("Getter del nombre")
        return self.__nombre

    @nombre.setter
    def nombre(self, nuevoNombre):
        print("Setter del codigo")
        self.__nombre = nuevoNombre

    @property
    def precio(self):
        print("Getter del precio")
        return self.__precio

    @precio.setter
    def precio(self, nuevoPrecio):
        print("Setter del Precio")
        self.__precio = nuevoPrecio

    @property
    def tipo(self):
        print("Getter del tipo")
        return self.__tipo

    @tipo.setter
    def tipo(self, nuevoTipo):
        print("Setter del tipo")
        self.__tipo = nuevoTipo

    def calcular_total(self, unidades):
        return self.__precio * unidades

    # El metodo __str__ esta programado para devolvernos siempre un string.
    def __str__(self):
        return " > Codigo: {}\n > Nombre: {}\n > Precio: {}\n > Tipo: {}".format(self.__codigo,self.__nombre,self.__precio,self.__tipo)

# Es mas seguro acceder por pasos a los datos que ir directamente a ellos, por eso los getters y setters. 
# El orden de la clase es: 1) Constructor, 2) Getters y setters, 3) Metodos propios, y 4) str. 
# OJO las indentaciones
# Las clases primeros y luego el main. 
if __name__ == "__main__":
    p1 = Producto(codigo = 1123,nombre = "Camisa",precio = 20,tipo = "Lino")
    print("--"*10)
    print("Usando el str al principio:")
    print(p1) # Es la invocacion del str
    print("--"*10)
    print("Usando los getters:")
    print(p1.codigo) # Getter del codigo
    print(p1.nombre) # Getter del nombre
    print(p1.precio) # Getter del precio
    print(p1.tipo) # Getter del tipo
    print("--"*10)
    print("Usando los setters:")
    p1.codigo = 1234 # Setter de codigo
    print(p1.codigo) # Getter de codigo
    p1.nombre = "Remera" # Setter de nombre
    print(p1.nombre) # Getter de nombre 
    p1.precio = 50 # Setter de precio
    print(p1.precio) # Getter de precio 
    p1.tipo = "Algodon" # Setter de tipo
    print(p1.tipo) # Getter de tipo 
    print("--"*10)
    print("Usando el str al final:")
    print(p1) # Es la invocacion del str
    print("--"*10)
    unidades5 = p1.calcular_total(5)
    print("Usando la funcion calcular_total:")
    print("Precio para el total de 5 unidades:", unidades5)
```

```python
# Seria una clase ABSTRACTA, como un concepto. 
class Atomos():

    def __init__(self,simbolo,nombre,electrones,numero_atomico):
        self.__simbolo = simbolo
        self.__nombre = nombre
        self.__electrones = electrones
        self.__numero_atomico = numero_atomico
        print("Se creo Atomos correctamente")

    @property 
    def simbolo(self):
        print("Getter del simbolo")
        return self.__simbolo

    @simbolo.setter
    def simbolo(self, nuevoSimbolo):
        print("Setter del simbolo")
        self.__simbolo = nuevoSimbolo

    @property
    def nombre(self):
        print("Getter del nombre")
        return self.__nombre

    @nombre.setter
    def nombre(self, nuevoNombre):
        print("Setter del codigo")
        self.__nombre = nuevoNombre

    @property
    def electrones(self):
        print("Getter de electrones")
        return self.__electrones

    @electrones.setter
    def electrones(self, nuevoElectrones):
        print("Setter de electrones")
        self.__electrones = nuevoElectrones

    @property
    def numero_atomico(self):
        print("Getter del tipo")
        return self.__numero_atomico

    @numero_atomico.setter
    def numero_atomico(self, nuevoNumero_atomico):
        print("Setter del tipo")
        self.__numero_atomico = nuevoNumero_atomico

    def perder_electron(self):
        self.__electrones -= 1
        return self.__electrones

    def ganar_electron(self):
        self.__electrones += 1
        return self.__electrones
    
    def ver_carga(self):
        return self.__electrones - self.__numero_atomico
    
    def __str__(self):
        return " > Simbolo: {}\n > Nombre: {}\n > Electrones: {}\n > Numero atomico: {}".format(self.__simbolo,self.__nombre,self.__electrones,self.__numero_atomico)


# Semimetales hereda de Atomos
class Semimetales(Atomos):

    def __init__(self,simbolo,nombre,electrones,numero_atomico,energia,conductor = False):
        super().__init__(simbolo,nombre,electrones,numero_atomico) # Atomos.__init__(self,simbolo,nombre,electrones,numero_atomico)
        self.__energia = energia
        self.__conductor = conductor 
        print("Se creo Semimetales correctamente")

    @property 
    def energia(self):
        print("Getter de energia")
        return self.__energia

    @energia.setter
    def energia(self, nuevaEnergia):
        print("Setter de energia")
        self.__energia = nuevaEnergia

    @property
    def conductor(self):
        print("Getter del conductor")
        return self.__conductor

    @conductor.setter
    def conductor(self, nuevoConductor):
        print("Setter del conductor")
        self.__conductor = nuevoConductor

    def encender_radiacion(self,energia_radiacion):
        if energia_radiacion > self.__energia:
            return self.__conductor == True
        else:
            return self.__conductor == False 
        
    def apagar_radiacion(self):
        self.__conductor == False
        return self.__conductor
    
    def __str__(self):
        return super().__str__() + "\n > Energia: {}\n > Conductor: {}".format(self.__energia,self.__conductor)


# Metales hereda de Atomos
class Metales(Atomos):

    def __init__(self,simbolo,nombre,electrones,numero_atomico,ultima_capa,conductor = True):
        super().__init__(simbolo,nombre,electrones,numero_atomico) # Atomos.__init__(self,simbolo,nombre,electrones,numero_atomico)
        self.__ultima_capa = ultima_capa
        self.__conductor = conductor
        print("Se creo Metales correctamente")

    @property
    def ultima_capa(self):
        print("Getter del ultima capa")
        return self.__ultima_capa

    @ultima_capa.setter
    def ultima_capa(self, nuevaUltima_capa):
        print("Setter del ultima capa")
        self.__ultima_capa = nuevaUltima_capa

    @property
    def conductor(self):
        print("Getter del conductor")
        return self.__conductor

    @conductor.setter
    def conductor(self, nuevoConductor):
        print("Setter del conductor")
        self.__conductor = nuevoConductor

    def magnetismo(self):
        if self.__ultima_capa.lower() == "d" or self.__ultima_capa.lower() == "f":
            return("Tiene magnetismo")
        else:
            return("No tiene magnetismo")

    def __str__(self):
        return super().__str__() + "\n > Ultima capa: {}\n > Conductor: {}".format(self.__ultima_capa,self.__conductor)


# Radioactivos hereda de Metales
class Radioactivos(Metales):
    
    def __init__(self,simbolo,nombre,electrones,numero_atomico,ultima_capa,conductor = True,radioactividad = True):
        super().__init__(simbolo,nombre,electrones,numero_atomico,ultima_capa,conductor = True) # Metales.__init__(self,simbolo,nombre,electrones,numero_atomico,ultima_capa,conductor = True)
        self.__radioactividad = radioactividad
        print("Se creo Radioactivos correctamente")

    @property
    def radioactividad(self):
        print("Getter de la radioactividad")
        return self.__radioactividad

    @radioactividad.setter
    def radioactividad(self, nuevaRadioactividad):
        print("Setter del radioactividad")
        self.__radioactividad = nuevaRadioactividad

    def ver_radioactividad(self):
        if self.__radioactividad == True:
            return ("Si.")
        else:
            return ("No.")

    def __str__(self):
        return super().__str__() + "\n > Radioactividad: {}".format(self.__radioactividad)


# NoRadioactivos hereda de Metales
class NoRadioactivos(Metales):
    
    def __init__(self,simbolo,nombre,electrones,numero_atomico,ultima_capa,conductor = True,radioactividad = False):
        super().__init__(simbolo,nombre,electrones,numero_atomico,ultima_capa,conductor = True) # Metales.__init__(self,simbolo,nombre,electrones,numero_atomico,ultima_capa,conductor = True)
        self.__radioactividad = radioactividad
        print("Se creo NoRadioactivos correctamente")

    @property
    def radioactividad(self):
        print("Getter de la radioactividad")
        return self.__radioactividad

    @radioactividad.setter
    def radioactividad(self, nuevaRadioactividad):
        print("Setter del radioactividad")
        self.__radioactividad = nuevaRadioactividad

    def ver_radioactividad(self):
        if self.__radioactividad == True:
            return ("Si.")
        else:
            return ("No.")

    def __str__(self):
        return super().__str__() + "\n > Radioactividad: {}".format(self.__radioactividad)


if __name__ == "__main__":
    print("=="*20)
    print("=="*20)
    print("Probando la clase Atomos")
    a1 = Atomos("H", "Hidrogeno", 1, 1)
   
    print("--"*20)
    print("Probando los getters de Atomos")
    print(a1.simbolo)
    print(a1.nombre)
    print(a1.electrones)
    print(a1.numero_atomico)
    
    print("--"*20)
    print("Probando str de Atomos:")
    print(a1)
   
    print("--"*20)
    print("Probando funciones de Atomos:")
    print("Perder un electron de a1:",a1.perder_electron()) 
    print("Ganar un electron de a1:",a1.ganar_electron()) 
    print("Ver carga de a1:",a1.ver_carga()) 
    
    print("--"*20)
    print("Probando los setters de Atomos")
    a1.simbolo = "X"
    print(a1.simbolo)
    a1.nombre = "Hola"
    print(a1.nombre)
    a1.electrones = 1
    print(a1.electrones)
    a1.numero_atomico = 1
    print(a1.numero_atomico)



    
    print("=="*20)
    print("=="*20)
    print("Probando la clase Semimetales:") # Pasa por los dos constructores
    sm1 = Semimetales(simbolo = "G",nombre = "Gera",electrones = 2,numero_atomico = 4,energia = 3,conductor = False) 
    
    print("--"*20)
    print("Probando los getters de Semimetales")
    print(sm1.simbolo)
    print(sm1.nombre)
    print(sm1.electrones)
    print(sm1.numero_atomico)
    print(sm1.energia)
    print(sm1.conductor)
    
    print("--"*20)
    print("Probando str de Semimetales")
    print(sm1)
    
    print("--"*20)
    print("Probando funciones de Semimetales:")
    print("Encendiendo la radiacion de sm1:",sm1.encender_radiacion(2))
    print("Apagando la radiacion de sm1:",sm1.apagar_radiacion())
    print("Ver carga de sm1:",sm1.ver_carga())
    
    print("--"*20)
    print("Probando los setters de Semimetales")
    sm1.simbolo = "X"
    print(sm1.simbolo)
    sm1.nombre = "Hola"
    print(sm1.nombre)
    sm1.electrones = 1
    print(sm1.electrones)
    sm1.numero_atomico = 1
    print(sm1.numero_atomico)
    sm1.energia = 1
    print(sm1.energia)
    sm1.conductor = True
    print(sm1.conductor)
   


    
    print("=="*20)
    print("=="*20)
    print("Probando la clase Metales") # Pasa por los dos constructores
    m1 = Metales("A", "Aron", 5, 8, "d", 3)
    
    print("--"*20)
    print("Probando los getters de Metales")
    print(m1.simbolo)
    print(m1.nombre)
    print(m1.electrones)
    print(m1.numero_atomico)
    print(m1.ultima_capa)
    print(m1.conductor)
    
    print("--"*20)
    print("Probando str de Metales:")
    print(m1)
    
    print("--"*20)
    print("Probando funciones de Metales:")
    print("Magnetismo de m1:",m1.magnetismo())
    
    print("--"*20)
    print("Probando los setters de Metales")
    m1.simbolo = "X"
    print(m1.simbolo)
    m1.nombre = "Hola"
    print(m1.nombre)
    m1.electrones = 1
    print(m1.electrones)
    m1.numero_atomico = 1
    print(m1.numero_atomico)
    m1.ultima_capa = 1
    print(m1.ultima_capa)
    m1.conductor = True
    print(m1.conductor)


    
    
    print("=="*20)
    print("=="*20)
    print("Probando la clase Radioactivos") # Pasa por los tres constructores
    r1 = Radioactivos("K", "Kaua", 10, 12, 14, True, False)
    
    print("--"*20)
    print("Probando los getters de Radioactivos")
    print(r1.simbolo)
    print(r1.nombre)
    print(r1.electrones)
    print(r1.numero_atomico)
    print(r1.ultima_capa)
    print(r1.conductor)
    print(r1.radioactividad)
    
    print("--"*20)
    print("Probando str de Radioactivos:")
    print(r1)
    
    print("--"*20)
    print("Probando funciones de Radioactivos:")
    print("Tiene radioactividad r1?",r1.ver_radioactividad())
    
    print("--"*20)
    print("Probando los setters de Radioactivos")
    r1.simbolo = "X"
    print(r1.simbolo)
    r1.nombre = "Hola"
    print(r1.nombre)
    r1.electrones = 1
    print(r1.electrones)
    r1.numero_atomico = 1
    print(r1.numero_atomico)
    r1.ultima_capa = 1
    print(r1.ultima_capa)
    r1.conductor = False
    print(r1.conductor)
    r1.radioactividad = True
    print(r1.radioactividad)



    
    print("=="*20)
    print("=="*20)
    print("Probando la clase NoRadioactivos") # Pasa por los tres constructores
    nr1 = NoRadioactivos("O", "Oros", 22, 44, 50, True, True)
    
    print("--"*20)
    print("Probando los getters de NoRadioactivos")
    print(nr1.simbolo)
    print(nr1.nombre)
    print(nr1.electrones)
    print(nr1.numero_atomico)
    print(nr1.ultima_capa)
    print(nr1.conductor)
    print(nr1.radioactividad)
    
    print("--"*20)
    print("Probando str de NoRadioactivos:")
    print(nr1)
    
    print("--"*20)
    print("Probando funciones de NoRadioactivos:")
    print("Tiene radioactividad nr1?",nr1.ver_radioactividad())
    
    print("--"*20)
    print("Probando los setters de NoRadioactivos")
    nr1.simbolo = "X"
    print(nr1.simbolo)
    nr1.nombre = "Hola"
    print(nr1.nombre)
    nr1.electrones = 1
    print(nr1.electrones)
    nr1.numero_atomico = 1
    print(nr1.numero_atomico)
    nr1.ultima_capa = 1
    print(nr1.ultima_capa)
    nr1.conductor = False
    print(nr1.conductor)
    nr1.radioactividad = False
    print(nr1.radioactividad)
    
```
