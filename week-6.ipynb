- Mostrar las 5 películas más antiguas del listado de títulos

- Mostrar las películas que en el titulo tienen la palabra "Dracula". También mostrar el número total de películas que coincidan con este requisito

```python
df_titulos_sNA = df_titulos.fillna(0)
dracula = df_titulos_sNA[df_titulos_sNA['title'].str.contains('Dracula', na=False)]
dracula.reset_index(drop=True)
# Son 83 peliculas que contienen la palabra Dracula 
```

- Mostrar los 10 títulos más comunes (que más se repiten)

```python
conteo_titulos = df_titulos['title'].value_counts()
conteo_titulos.head(10)
```

- Mostrar cual fue la primera película hecha titulada "Romeo and Juliet"

```python
romeo_and_juliet = df_titulos[ df_titulos ["title"] == "Romeo and Juliet"]
romeo_and_juliet_ordenado = romeo_and_juliet.sort_values(by = ["year"], ascending= True)
romeo_and_juliet_ordenado.head(1)
```

- Listar todas las películas que contengan la palabra "Exorcist" ordenadas de la más vieja a la más reciente

```python
exorcist = df_titulos[df_titulos['title'].str.contains('Exorcist', na=False)]
exorcist_ordenados = exorcist.sort_values('year')
exorcist_ordenados.reset_index(drop=True)
```

- Mostrar cuantas películas fueron hechas en el año 1950

```python
peliculas1950 = df_titulos[df_titulos ["year"] == 1950]
peliculas1950.reset_index(drop=True)
```

- Mostrar cuantas películas fueron hechas de 1950 a 1959

```python
condicion_1950_1959 = (df_titulos["year"] >= 1950) & (df_titulos["year"] <= 1959)
peliculas1950_1959 = df_titulos.loc[condicion_1950_1959]
peliculas1950_1959_ordenadas = peliculas1950_1959.sort_values(by = ["year"], ascending = True)
peliculas1950_1959_ordenadas.reset_index(drop=True)
```

- Mostrar el elenco completo ordenado por la clasificación "n" de la película "Dracula" de 1958

```python
dracula = df_elenco.loc[df_elenco["title"] == "Dracula"] # Todas las peliculas de dracula
dracula = dracula.fillna(0) # Relleno Nan con 0 
dracula_1958 = dracula.loc[dracula["year"] == 1958] # Dracula de 1958
dracula_1958.sort_values(by = ["n"], ascending = True) # Ordeno de menor a mayor por n
```

- Mostrar cuantos papeles ha hecho "Robert De Niro" en su carrera

- Listado de papeles como protagonista (n=1) que tuvo el actor "Charlton Heston" en la década de los 60's, ordenado por año de forma descendente

```python
n1 = df_elenco[df_elenco["n"] == 1]
charlton_heston_n1 = n1.loc[n1["name"] == "Charlton Heston"]
condicion_60 = (charlton_heston_n1["year"] >= 1960) & (charlton_heston_n1["year"] <= 1969)
charlton_heston_n1_60 = charlton_heston_n1[condicion_60]
charlton_heston_n1_60.sort_values("year")
```
